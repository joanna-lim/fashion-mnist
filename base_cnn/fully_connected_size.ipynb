# Hyperparameter Tuning: Fully Connected Layer Size

```python
import torch
import torch.nn as nn
from torchvision import datasets, transforms

transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])

train_dataset = datasets.FashionMNIST(root='../data', train=True, download=True, transform=transform)
test_dataset = datasets.FashionMNIST(root='../data', train=False, download=True, transform=transform)
```

#### Run Test Harness
Perform CNN on varying sizes: [128, 256, 512]

```python
from test_harness import run_test_harness

# control variables:
batch_size = 32
activation_fn = nn.ReLU()

fc_layer_sizes = [128, 256, 512]
accuracies = []
for fc_layer_size in fc_layer_sizes:
    print(f"---Fully Connected Layer Size: {fc_layer_size}---")
    accuracy = run_test_harness(train_dataset, test_dataset, batch_size, fc_layer_size, activation_fn)
    accuracies.append((fc_layer_size, accuracy))
```

```python
import matplotlib.pyplot as plt

fc_size, accuracy = zip(*accuracies) 

plt.figure(figsize=(8, 6))
plt.plot(fc_layer_sizes, accuracy, marker='o', linestyle='-', label='Accuracy')
plt.title('Accuracy vs. Fully Connected Layer Sizes')
plt.xlabel('Fully Connected Layer Sizes')
plt.ylabel('Accuracy')
plt.show()
```

#### Optimal Fully Connected Layer Size: 256
