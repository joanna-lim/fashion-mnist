# 1. Hyperparameter Tuning for CNN: Batch Size

```python
import torch
import torch.nn as nn
from torchvision import datasets, transforms
```

#### Getting datasets

```python
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])

train_dataset = datasets.FashionMNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.FashionMNIST(root='./data', train=False, download=True, transform=transform)
```

#### Run Test Harness
Perform CNN with varying batch sizes: [32, 64, 128]




```python
from test_harness import run_test_harness
# control variables:
    # size of fully connected layers: 128
    # choice of activation function: ReLU
    # number of epochs = 10

batch_sizes = [32, 64, 128]

accuracies = []
for batch_size in batch_sizes:
    print("---batch size: {}---".format(batch_size))
    accuracy = run_test_harness(train_dataset, test_dataset, batch_size=batch_size, fc_layer_size=128, activation_fn=nn.ReLU(), num_epochs=10)
    accuracies.append((batch_size, accuracy))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
batch_sizes, accs = zip(*accuracies)
plt.plot(batch_sizes, accs, marker='o')
plt.title('Accuracy vs. Batch Size')
plt.xlabel('Batch Size')
plt.ylabel('Accuracy')
plt.show()
```

#### Optimal batch size: 32.
