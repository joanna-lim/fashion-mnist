# CNN without MixUp vs CNN with MixUp

```python
import matplotlib.pyplot as plt

# Experiment results
base_cnn = [85.67, 88.39, 89.55, 90.32, 90.3, 90.37, 89.45, 91.08, 91.52, 91.09, 91.52, 91.07]
cnn_with_mixup = [85.84, 87.23, 88.45, 89.0, 89.45, 89.89, 90.51, 90.87, 90.72, 90.85, 91.01, 91.03, 91.51, 91.67, 91.34, 91.79, 91.81, 91.89, 92.13, 92.16]

# Adjust the data for the base_cnn to match the epochs of cnn_with_mixup
epochs_base_cnn = list(range(1, len(base_cnn) + 1))
epochs_cnn_with_mixup = list(range(1, len(cnn_with_mixup) + 1))

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(epochs_base_cnn, base_cnn, marker='o', label='Base CNN with Early Stopping')
plt.plot(epochs_cnn_with_mixup, cnn_with_mixup, marker='o', label='CNN with Mixup')

# Adding text annotation for early stopping
x_coord = len(base_cnn)
y_coord = base_cnn[-1]
plt.annotate('Early Stopping', xy=(x_coord, y_coord), xytext=(x_coord-2, y_coord-2),
             arrowprops=dict(facecolor='black', shrink=0.05, width=0.5, headwidth=8))

plt.title('Performance Comparison of Base CNN and CNN with Mixup')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
