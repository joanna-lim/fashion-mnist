# Comparing Across Different Models

```python
import matplotlib.pyplot as plt

models = ['CNN', 'CNN with Mixup', 'ViT', 'ViT with Mixup']
accuracies = [91.52, 92.16, 92.3, 92.32]
training_times = ['0:03:31.073759', '0:05:50.671932', '3:12:07.490904', '3:12:11.459586']

# convert training times to total seconds
def time_to_sec(time_str):
    h, m, s = map(float, time_str.split(':'))
    return 3600 * h + 60 * m + s

training_seconds = [time_to_sec(time) for time in training_times]

fig, ax1 = plt.subplots(figsize=(10, 6))

# plotting accuracy
ax1.bar(models, accuracies, color='skyblue')
ax1.set_xlabel('Models')
ax1.set_ylabel('Accuracy', color='skyblue')
ax1.tick_params('y', colors='skyblue')
ax1.set_ylim(91, 93)  

# creating a secondary y-axis for training time
ax2 = ax1.twinx()
ax2.plot(models, training_seconds, marker='o', color='lightcoral', linestyle='dashed', linewidth=2, markersize=8)
ax2.set_ylabel('Training Time (seconds)', color='salmon')
ax2.tick_params('y', colors='salmon')

plt.title('Model Comparison Based on Accuracy and Training Time')
plt.show()
```
